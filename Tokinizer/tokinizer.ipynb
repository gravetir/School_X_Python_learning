```python
def lexical_text(text_1):
    with open(text_1, 'r', encoding="utf-8") as file:
        text = file.read()

    tokens = text.split()
   
    unique_words = set(tokens)
    num_unique_words = len(unique_words)
    
    lexical_text = num_unique_words / len(tokens)

    return lexical_text

text_1 = "141752_litres.txt"  
diversity_score = lexical_text(text_1)
print(diversity_score)
```

```python
import string

class Tokenizer:
    @staticmethod
    def tokenizer(text):
# Удаление знаков препинания из текста
        translator = str.maketrans('', '', string.punctuation)
        text = text.translate(translator)

# Разделение текста на токены по пробелам
        tokens = text.split()

        return tokens

```

```python
# Пример использования
text = "Идёт человек по улице, а за ним другие: рабочие, дворяне, крестьяне!!!!"
tokenizer = Tokenizer()
tokens = tokenizer.tokenizer(text)
print(tokens)
```
