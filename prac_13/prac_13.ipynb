```python
import numpy as np
import random


def flip_dN(N: int=2, times: int=1) -> list[int]:
    return [random.randint(1,N) for _ in range(times)]
```

```python
flip_dN(20,3)
```

```python
times= 5000
d2_flips=flip_dN(2,times)
d20_flips=flip_dN(20,times//10)
d2_flips, d20_flips
```

```python
np.mean(d2_flips)
```

```python
np.mean(d20_flips)
```
